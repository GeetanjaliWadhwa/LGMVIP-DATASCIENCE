**DATA WRANGLING/FEATURE ENGG**

```python
iris['class']=iris['species'].str.split("-", expand=True)[1]
iris.head()

#[1] represents the index of string after split eg, iris=["iris","setosa"]
#expand=True: return dataframe/multiindex expanding dimentionality (split the strings intonseparate columns)
```

```python
iris.drop('species',axis=1,inplace= True)
iris.head()

#inpace: False returns a copy of existing database
```

**TEST TRAIN ALGORITHM**

```python
X=iris[['sepal length','sepal width','petal length','petal width']]
y= iris['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, stratify=y,shuffle=True)

#data is split in a stratified fashion, using this as the class labels
```

```python
index=["X_train", "X_test", "y_train", "y_test"]
count=0
for i in [X_train, X_test, y_train, y_test]:
  col_name = index[0+count]
  count+=1
  print( f"length of {col_name} = ", len(i))
```

**CHOOSING THE RIGHT CLASSIFIER & MODEL SELECTION**

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn.linear_model import LogisticRegression

from sklearn import metrics

from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve, auc
```

```python
def cf_model(model1):
   model=model1()
   model.fit(X_train,y_train)
   predicted = model.predict(X_test)
   expected= y_test

   print(classification_report(expected, predicted))

   cm = confusion_matrix(expected,predicted)
   print ("Confusion Matrix:\n",cm)
   labels=['setosa','versicolor','virginica']

   df_matrix= pd.DataFrame(cm, index=labels, columns=labels)

   sb.heatmap(df_matrix, annot=True, fmt='d')
   plt.title("Confusion Matrix")
   plt.xlabel("Predicted Label")
   plt.ylabel("Expected label")
   plt.show()

   print("Accuracy= ", accuracy_score(expected,predicted))
```

```python
cf_model(AdaBoostClassifier)
```

```python
cf_model(GaussianNB)
```

```python
cf_model(LogisticRegression)
```

```python
cf_model(KNeighborsClassifier)
```

```python
cf_model(SVC)
```

```python
cf_model(DecisionTreeClassifier)
```
